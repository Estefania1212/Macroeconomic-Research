```python
import matplotlib.pyplot as plt
import pandas as pd

# Define the event dates and labels
event_dates = [
    '2018-07-06', '2019-05-10', '2020-03-27', '2020-12-27',
    '2021-03-11', '2021-11-15', '2022-08-16', '2023-12-01', '2025-04-02'
]
event_labels = [
    'Tariffs 1', 'Tariffs 2', 'CARES Act', 'COVID Relief 2',
    'Rescue Plan', 'Infrastructure', 'Inflation Reduction Act', 'Omnibus', 'Liberation Day'
]
event_dates = pd.to_datetime(event_dates)

# Create the plot
fig, ax = plt.subplots(figsize=(14, 7))
ax.plot(filtered_data['Date'], filtered_data['USD'], label='USD Exchange Rate', color='blue')

# Add vertical lines and annotations using axes coordinates
for date, label in zip(event_dates, event_labels):
    ax.axvline(date, color='red', linestyle='--', linewidth=1.5)
    ax.annotate(label,
                xy=(date, 1.01),
                xycoords=('data', 'axes fraction'),
                rotation=90,
                ha='center',
                va='bottom',
                fontsize=11,
                color='black')

# Title and final formatting
plt.suptitle('USD Exchange Rate vs Fiscal Policy Events (2017–2025)', fontsize=15, y=1.02)
ax.set_xlabel('Date')
ax.set_ylabel('USD Value')
ax.legend()
ax.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Analyze reaction of USD to each fiscal policy event
reaction_results = []

for date, label in zip(event_dates, event_labels):
    # Define a small window before and after the event
    before = filtered_data[(filtered_data['Date'] >= date - pd.Timedelta(days=5)) & 
                           (filtered_data['Date'] < date)]
    after = filtered_data[(filtered_data['Date'] > date) & 
                          (filtered_data['Date'] <= date + pd.Timedelta(days=5))]
    
    # Calculate average rate before and after the event
    before_rate = before['USD'].mean()
    after_rate = after['USD'].mean()
    
    if pd.notnull(before_rate) and pd.notnull(after_rate):
        change = after_rate - before_rate
        direction = "Appreciated" if change < 0 else "Depreciated"
        reaction_results.append((label, before_rate, after_rate, round(change, 4), direction))
```

```python
# Create a DataFrame to summarize the reactions
summary_df = pd.DataFrame(reaction_results, columns=[
    'Event', 'Rate Before', 'Rate After', 'Change', 'Direction'
])

print(summary_df)


#EUR/USD ↑ = USD depreciates (you need more USD to buy 1 EUR)

#EUR/USD ↓ = USD appreciates (you need fewer USD to buy 1 EUR)

#So:

#If Rate After > Rate Before → USD Depreciated

#If Rate After < Rate Before → USD Appreciated
```

```python

#Following the implementation of large fiscal stimulus packages such as the CARES Act (2020) and the American Rescue Plan (2021), the U.S. dollar experienced noticeable depreciation. This occurred because the government injected substantial amounts of money into the economy, increasing the money supply. With more dollars in circulation and continued low interest rates, the value of the dollar weakened against other currencies. As more people had cash to spend, demand for goods and services surged, but supply chains were still strained. This mismatch between high demand and limited supply drove prices upward, leading to a sharp increase in inflation, particularly in 2021 and 2022. The graph clearly illustrates this trend, showing the decline in USD value following the stimulus events and a corresponding rise in inflation during the same period.
#However, this relationship is not always linear. If fiscal stimulus is accompanied by tight monetary policy, such as higher interest rates, the inflationary impact can be contained. Higher rates make the dollar more attractive to investors, supporting its value and dampening domestic demand. This balancing act between fiscal expansion and monetary restraint can prevent inflation from spiraling while maintaining currency stability.
#In contrast to earlier stimulus packages, later fiscal policies such as the Infrastructure Investment and Jobs Act (2021) and the Inflation Reduction Act (2022) were followed by a period of USD appreciation, as seen in the graph. Although these policies still involved significant government spending, the macroeconomic context had changed. By this time, inflation had already surged, and the Federal Reserve began raising interest rates aggressively in response. These higher interest rates made the U.S. dollar more attractive to foreign investors, increasing capital inflows and strengthening the currency.

#As a result, despite ongoing fiscal expansion, the dollar appreciated — demonstrating how tight monetary policy can counteract inflationary and depreciative effects of government spending. This reflects the important balance between fiscal and monetary policy in shaping exchange rate behavior and controlling inflation.
#Interest rates are the main tool used by central banks to control inflation. When inflation rises too quickly, central banks raise interest rates to slow down borrowing and reduce demand. This helps ease pressure on prices. On the other hand, when inflation is low or the economy is weak, they lower rates to encourage borrowing and spending.
```

```python
import matplotlib.pyplot as plt
import matplotlib.lines as mlines
import matplotlib.cm as cm
import numpy as np
import pandas as pd
from collections import defaultdict

# === Filter Plot Data ===
plot_data = merged_cpi[(merged_cpi['Year'] >= 2017) & (merged_cpi['Year'] <= 2025)]

# === Setup Plot ===
fig, ax1 = plt.subplots(figsize=(17, 9))
ax1.plot(plot_data['Year'], plot_data['Avg_USD'], color='blue', label='USD Exchange Rate')
ax1.set_xlabel('Year')
ax1.set_ylabel('USD Exchange Rate', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')

# Inflation Plot
ax2 = ax1.twinx()
ax2.plot(plot_data['Year'], plot_data['Inflation (%)'], color='orange', label='European Inflation (%)')
ax2.set_ylabel('European Inflation (%)', color='orange')
ax2.tick_params(axis='y', labelcolor='orange')

# === Event Setup ===
summary_df['Event_Date'] = pd.to_datetime([
    '2018-07-06', '2019-05-10', '2020-03-27', '2020-12-27',
    '2021-03-11', '2021-11-15', '2022-08-16', '2023-12-01', '2025-04-02'
])
summary_df['Event_Year'] = summary_df['Event_Date'].dt.year

# Colors and arrows
cmap = plt.colormaps['tab10']
colors = [cmap(i % cmap.N) for i in range(len(summary_df))]
arrow_length = 0.03
head_width = 0.07
head_length = 0.012

# Event spacing for jitter
event_spacing = defaultdict(list)
for idx, row in summary_df.iterrows():
    year = row['Event_Year']
    event_spacing[year].append(idx)

# Legend markers
event_markers = []

# === Plot Events Without Labels ===
for i, row in summary_df.iterrows():
    year = row['Event_Year']
    color = colors[i]

    # Jitter for same-year events
    position = event_spacing[year].index(i)
    total = len(event_spacing[year])
    spacing_factor = 0.18
    offset = (position - (total - 1) / 2) * spacing_factor
    year_jittered = year + offset

    # Only one vertical line per year
    if position == 0:
        ax1.axvline(year, color='gray', linestyle='--', linewidth=1)

    # USD data
    nearest_index = (plot_data['Year'] - year).abs().idxmin()
    usd_value = plot_data.loc[nearest_index, 'Avg_USD']

    # Dot (event)
    ax1.plot(year_jittered, usd_value, 'o', color=color, markersize=10)

    # Arrow directly aligned to dot
    arrow_color = 'red' if row['Direction'] == 'Depreciated' else 'green'
    arrow_y_start = usd_value - arrow_length / 2 if arrow_color == 'red' else usd_value + arrow_length / 2

    ax1.arrow(year_jittered, arrow_y_start, 0,
              arrow_length if arrow_color == 'red' else -arrow_length,
              color=arrow_color, head_width=head_width, head_length=head_length,
              length_includes_head=True, alpha=0.85)

    # Legend entry
    legend_item = mlines.Line2D([], [], color=color, marker='o', linestyle='None',
                                markersize=10, label=row['Event'])
    event_markers.append(legend_item)

# === Final Formatting ===
ax1.legend(handles=event_markers, loc='upper center', bbox_to_anchor=(0.3, -0.5),
           fancybox=True, shadow=True, ncol=5, fontsize=13, title="Fiscal Events")

plt.subplots_adjust(bottom=0.35)
plt.title('IMPACT OF FISCAL POLICY ON USD AND EUROPEAN INFLATION (2017–2025)', fontsize=16, weight='bold')
fig.tight_layout()
plt.grid(True)
plt.show()
```
